```r
## feature engineering
```

```r
data1$volatility <- (data$Adj..High-data$Adj..Close)/data$Adj..Close
data1$PCT_Change <- (data$Adj..Close-data$Adj..Open)/data$Adj..Open
```

```r
data1$Adj..Close = as.factor(data1$Adj..Close)
```
